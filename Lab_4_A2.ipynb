```
#A2
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

# Load the Excel file
xls = pd.ExcelFile("Lab Session Data.xlsx")

# Load the correct sheet
df = xls.parse("IRCTC Stock Price")

# Clean column names
df.columns = df.columns.str.strip()

# Define corrected required columns
required_cols = ["Open", "High", "Low", "Price"]
df = df.dropna(subset=required_cols)

# Define features (X) and target (y)
X = df[["Open", "High", "Low"]]
y = df["Price"]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display results
print(f"\nModel Evaluation Metrics:")
print(f"MSE (Mean Squared Error): {mse:.2f}")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f}")
print(f"MAPE (Mean Absolute Percentage Error): {mape * 100:.2f}%")
print(f"R² Score (Coefficient of Determination): {r2:.4f}")
```

```
#A2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load both Excel files

df1 = pd.read_csv('/content/ptbxl_database.csv')
df2 = pd.read_csv('/content/scp_statements.csv')

# Combine the datasets
df_combined = pd.concat([df1, df2], ignore_index=True)

# Select the feature for A2 (e.g., 'age')
feature = 'age'

# Drop missing values in the selected feature
feature_data = df_combined[feature].dropna()

# Filter out unrealistic ages (e.g., <0 or >120)
clean_age_data = feature_data[(feature_data >= 0) & (feature_data <= 120)]

# Calculate mean and variance of cleaned data
mean_clean = np.mean(clean_age_data)
var_clean = np.var(clean_age_data)

print(f"Cleaned Mean of '{feature}': {mean_clean:.2f}")
print(f"Cleaned Variance of '{feature}': {var_clean:.2f}")

# Plot histogram using matplotlib
plt.figure(figsize=(8, 5))
plt.hist(clean_age_data, bins=10, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title(f"Cleaned Histogram of {feature} (0–120 years)")
plt.xlabel(feature)
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
# A2:
import pandas as pd
import ast
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error
import numpy as np

# -------------------------------------------------------
# Function to load ECG data
# -------------------------------------------------------
def load_ecg_data_for_regression(ecg_csv_path, scp_csv_path):
    """
    Loads PTB-XL metadata, selects numeric features, and returns DataFrame.
    """
    ecg_df = pd.read_csv(ecg_csv_path)
    return ecg_df

# ------------------------------------------------
# Function to preprocess data
# ------------------------------------------------
def preprocess_regression_data(df, feature_columns, target_column, test_size=0.3, random_state=42):
    """
    Prepares features and target arrays for regression.
    Handles missing values, scales features, and splits into train/test sets.
    """
    X = df[feature_columns]
    y = df[target_column]

    # Impute missing values
    imputer = SimpleImputer(strategy='mean')
    X_imputed = imputer.fit_transform(X)

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_imputed)

    # Split into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=random_state
    )

    return X_train, X_test, y_train, y_test

# ---------------------------------------
# Function to train Linear Regression
# ---------------------------------------
def train_linear_regression(X_train, y_train):
    """
    Trains a Linear Regression model and returns it.
    """
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model

# ------------------------------------
# Function to evaluate regression model
# ------------------------------------
def evaluate_regression_model(model, X_train, y_train, X_test, y_test):
    """
    Calculates MSE, RMSE, MAPE, and R2 for train and test sets.
    """
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Metrics
    metrics = {}
    for dataset, y_true, y_pred in [
        ('train', y_train, y_train_pred),
        ('test', y_test, y_test_pred)
    ]:
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        mape = mean_absolute_percentage_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)

        metrics[dataset] = {
            'MSE': mse,
            'RMSE': rmse,
            'MAPE': mape,
            'R2': r2
        }
    return metrics

# -----------------
# Main program
# -----------------
if __name__ == "__main__":
    # Step 1: Load data
    ecg_data = load_ecg_data_for_regression(
        "ptbxl_database.csv",
        "scp_statements.csv"
    )

    # Step 2: Define features and target (numeric target for regression)
    feature_columns = ['height', 'weight']  # predictors
    target_column = 'age'                   # numeric target

    # Step 3: Preprocess data
    X_train, X_test, y_train, y_test = preprocess_regression_data(
        ecg_data, feature_columns, target_column
    )

    # Step 4: Train model
    lr_model = train_linear_regression(X_train, y_train)

    # Step 5: Evaluate model
    metrics_result = evaluate_regression_model(
        lr_model, X_train, y_train, X_test, y_test
    )

    # Step 6: Output results
    print("A2 Result → Regression Metrics")
    for dataset in metrics_result:
        print(f"\n{dataset.upper()} SET:")
        for metric_name, value in metrics_result[dataset].items():
            print(f"{metric_name}: {value:.4f}")
```

```
#A3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate 20 random (X, Y) coordinates between 1 and 10
X = np.random.uniform(1, 10, size=(20, 2))

# Create a DataFrame
df = pd.DataFrame(X, columns=["X", "Y"])

# Simple rule: if X + Y > 10 â†’ Class 1 (Red), else Class 0 (Blue)
df["Class"] = np.where(df["X"] + df["Y"] > 10, 1, 0)

# Assign colors based on class
df["Color"] = df["Class"].map({0: "blue", 1: "red"})

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df["X"], df["Y"], c=df["Color"], s=100, edgecolor="black")
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.title("Scatter Plot of 20 Training Points")
plt.grid(True)
plt.show()
```
